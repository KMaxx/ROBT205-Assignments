```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2.1
def max_lab(v1, v2, v3):
    if type(v1) != int and type(v1) != float or type(v2) != int and type(v2) != float or type(v3) != int and type(v3) != float:
        raise Exception("The type of the inputted value is wrong!")
    else:
        return max(v1, v2, v3)
```

```python
#2.2
def sum_lab(l):
    sum_l = 0
    for i in l:
        sum_l = sum_l + i
    return sum_l

def mult_lab(l):
    mult_l = 1
    for i in l:
        mult_l = mult_l*i
    return mult_l
```

```python
#2.3
def fact_lab(v):
    if v<0 or type(v) != int:
        raise Exception("The type of the inputted value is wrong!")
    fact_l = 1
    for i in np.arange(1,v+1):
        fact_l = fact_l*i
    return fact_l
```

```python
#2.4
def zeros_lab(r, c):
    return np.zeros([r,c])
```

```python
#2.5
def identity_lab(n):
    m = zeros_lab(n,n)
    for i in range(n):
        m[i][i] = 1
    return m
```

```python
#2.6
def convolve_lab(l1,l2):
    if np.size(l1) < np.size(l2):
        k = l1
        m = l2
    else:
        k = l2
        m = l1
    o = np.zeros([np.size(k), np.size(m)+np.size(k)-1])
    t = 0
    for i in m:
        o = o + i*np.eye(np.size(k), np.size(m)+np.size(k)-1, t)
        t=t+1
    return np.matmul(k,o)
```

```python
convolve_lab([0,0,0,4,4,5,6,6,6,5,4,3,2,1,0,8,0],[-1,2,-1])
```

```python
#2.7
import random

def rand255(r, c):
    m = np.array([random.randint(0, 255) for i in range(r*c)])
    return m.reshape(r,c)
```

```python
#2.8
R = np.pad(rand255(21,21), pad_width=1,mode='constant', constant_values=0)
M = np.zeros([3,3])-1
M[1,1] = 8
O = np.zeros([21,21])
for i in np.arange(1, 22):
    for j in np.arange(1, 22):
        O[i-1][j-1] = sum(sum(M*R[i-1:i+2, j-1:j+2]))
O
```

```python
def mcl(x, n):
    a = np.zeros(n)
    for i in range(n):
        if i == 0:
            a[i] = 1
            continue
        a[i] = a[i-1] + x**i
    return a
```

```python
j=0.999
val = mcl(j, 10000)
err = 1/(1-j) + np.zeros(np.size(val)) - val
plt.plot(range(np.size(err)),err)
plt.xlabel("Number of terms")
plt.ylabel("Truncated error")
```
